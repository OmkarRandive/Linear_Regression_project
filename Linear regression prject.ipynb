```python
df_num=df[["age","bmi","children","charges"]]
```

```python
df_num
```

```python
#create a list of categorical variables.
df.select_dtypes(include=["object"]).columns.tolist()
```

```python
#convert sex column to numerical data using get_dummies
df_dummies1=pd.get_dummies(df["sex"])
```

```python
df_dummies1.head()
```

```python
#convert smoker column to numerical data using get_dummies
df_dummies2=pd.get_dummies(df["smoker"])
```

```python
df_dummies2.head()
```

```python
#convert region column to numerical data using get_dummies
df_dummies3=pd.get_dummies(df["region"])
```

```python
df_dummies3.head()
```

```python
#combined column wise , appending of columns
df_combined=pd.concat([df_num,df_dummies1,df_dummies2,df_dummies3],axis=1)
```

```python
df_combined.head(5)
```

```python
#find missing values
df_combined.isnull().sum(axis=0)
```

```python
x=df_combined.drop("charges",axis=1)
```

```python
y=df_combined["charges"]
```

```python
x.head()
```

```python
y.head()
```

```python
#splitting the data into training set and testing set. splitting 80% into training set and 30% into testing set
```

```python
#split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=12)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
#standardize or feature scalling the dataset.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
#performing linear regression.
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm
```

```python
lm.fit(x_train,y_train)
```

```python
#model evaluation
```

```python
#print the intercept
print(lm.intercept_)
```

```python
#print the coefficient
print(lm.coef_)
```

```python
#making predictions using the model
y_pred=lm.predict(x_test)
y_pred
```

```python
#relationship between real data and predicted data
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Real data")
plt.ylabel("predicted data")
```

```python
#calculating residuals
```

```python
residuals=y_test-y_pred
```

```python
residuals
```

```python
#distribution of residuals are normal distribution
sns.displot(residuals,kind="kde")
```

```python
#scatterplot with predictions and residuals.
plt.scatter(y_test,residuals)
```

```python
#performance metrics
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
#R squared and adjusted R squared
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
#adjusted R squared
1-(1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
